## Creating an image vector


```python
import os
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
```

```python
image_dir = os.path.expanduser('~/projects/tf20_vision_samples/cv_images')
print (image_dir)
```

```python
image_path = os.path.join(image_dir, "16076197128-1-0-c.jpg")
image = Image.open(image_path)
plt.figure(figsize=(8, 8))
plt.imshow(np.asarray(image))
```

```python
image_arr = np.asarray(image)
image_arr.shape
```

```python
gray = cv2.cvtColor(image_arr, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8, 8))
plt.imshow(gray, cmap='gray')
```

```python
# RED channel

plt.figure(figsize=(8, 8))
plt.imshow(image_arr[:,:, 0])
```

```python
# GREEN channel

plt.figure(figsize=(8, 8))
plt.imshow(image_arr[:,:,1])
```

```python
# BLUE channel

plt.figure(figsize=(8, 8))
plt.imshow(image_arr[:,:,2])
```
