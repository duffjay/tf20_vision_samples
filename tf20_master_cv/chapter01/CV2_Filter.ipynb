```python
import os
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
```

```python
image_dir = os.path.expanduser('~/projects/tf20_vision_samples/cv_images')
print (image_dir)
```

```python
image_path = os.path.join(image_dir, "16076197128-1-0-c.jpg")
image = Image.open(image_path)
plt.figure(figsize=(8, 8))
plt.imshow(np.asarray(image))
```

```python
image_arr = np.asarray(image)
image_arr.shape
```

```python
gray = cv2.cvtColor(image_arr, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(8, 8))
plt.imshow(gray, cmap='gray')
```

```python
kernel_horiz = np.array([[-1,-1,-1], [ 2, 2, 2], [-1,-1,-1]])
kernel_vert =  np.array([[-1, 2,-1], [-1, 2,-1], [-1, 2,-1]])
kernel_corner_1 = np.array([[ 2, 2, 2], [ 2,-1,-1], [ 2,-1,-1]])
```

```python
filter_horiz = cv2.filter2D(gray, -1, kernel_horiz)
plt.figure(figsize=(8, 8))
plt.imshow(filter_horiz, cmap='gray')
```

```python
filter_vert = cv2.filter2D(gray, -1, kernel_vert)
plt.figure(figsize=(8, 8))
plt.imshow(filter_vert, cmap='gray')
```

```python
filter_corner_1 = cv2.filter2D(gray, -1, kernel_corner_1)
plt.figure(figsize=(8, 8))
plt.imshow(filter_corner_1, cmap='gray')
```
