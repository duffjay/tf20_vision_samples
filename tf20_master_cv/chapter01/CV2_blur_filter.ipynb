```python
from __future__ import print_function
import os
import numpy as np
import cv2
from skimage.util import random_noise

import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
```

```python
image_dir = os.path.expanduser('~/projects/tf20_vision_samples/cv_images')
print (image_dir)
```

```python
image_path = os.path.join(image_dir, "16076197128-1-0-c.jpg")
image = Image.open(image_path)
plt.figure(figsize=(8, 8))
plt.imshow(np.asarray(image))
```

```python
image_arr = np.asarray(image)
image_arr.shape
```

```python
gray = cv2.cvtColor(image_arr, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(8, 8))
plt.imshow(gray, cmap='gray')
```

```python
imgnoise = random_noise(image_arr, mode='s&p', amount=0.1)
plt.figure(figsize=(8, 8))
plt.imshow(imgnoise, cmap='gray')
```

## Mean

```python
# 3,3 vs 5,5
# - increases the blur
meanimg = cv2.blur(image_arr, (5,5))
plt.figure(figsize=(8, 8))
plt.imshow(meanimg, cmap='gray')
```

```python
medianimg = cv2.medianBlur(image_arr,5)
plt.figure(figsize=(8, 8))
plt.imshow(medianimg, cmap='gray')
```

```python
gaussianimg = cv2.GaussianBlur(image_arr, (5,5),0)
plt.figure(figsize=(8, 8))
plt.imshow(gaussianimg, cmap='gray')
```
